```python
import rasterio
from rasterio.plot import show
import matplotlib.pyplot as plt

def view_raster(raster_path):
    # Open the raster file
    raster = rasterio.open(raster_path)

    # Display the raster
    show(raster, cmap='viridis')

    # Add zoom and pan functionalities
    plt.show()

view_raster(r"C:\Users\Rohit Verma\Downloads\NE1_LR_LC\NE1_LR_LC.tif")
```

```python
import rasterio
from rasterio.plot import show
import matplotlib.pyplot as plt

def change_band_combination(raster_path, bands):
    # Open the raster file
    raster = rasterio.open(raster_path)

    # Display the raster with the specified bands
    show((raster, *bands), cmap='viridis')

    # Add zoom and pan functionalities
    plt.show()

change_band_combination(r"C:\Users\Rohit Verma\Downloads\NE1_LR_LC\NE1_LR_LC.tif", bands=(1, 2, 3))
```

```python
import geopandas as gpd
import matplotlib.pyplot as plt

def plot_shapefile_attribute_chart(shapefile_path, attribute_column):
    # Read the shapefile
    gdf = gpd.read_file(shapefile_path)

    # Print available attribute columns
    print("Available attribute columns:", gdf.columns)

    # Plot a bar chart of the specified attribute data
    gdf[attribute_column].plot(kind='bar', legend=True)

    # Show the chart
    plt.show()

# Example usage
plot_shapefile_attribute_chart(r"C:\Users\Rohit Verma\Downloads\antartica\gis_osm_pois_a_free_1.shp", attribute_column='population')
```

```python
import rasterio
import numpy as np
from rasterio.plot import show
import matplotlib.pyplot as plt

def numpy_array_manipulation(raster_path):
    # Open the raster file
    raster = rasterio.open(raster_path)

    # Read raster data as a NumPy array
    raster_array = raster.read(1)

    # Perform NumPy operations (e.g., mathematical transformations, filtering)
    manipulated_array = np.sqrt(raster_array)

    # Display the original and manipulated arrays
    fig, ax = plt.subplots(1, 2, figsize=(10, 5))

    ax[0].imshow(raster_array, cmap='viridis')
    ax[0].set_title('Original Raster')

    ax[1].imshow(manipulated_array, cmap='viridis')
    ax[1].set_title('Manipulated Raster')

    plt.show()

# Example usage
numpy_array_manipulation(r"C:\Users\Rohit Verma\Downloads\NE1_LR_LC\NE1_LR_LC.tif")
```
